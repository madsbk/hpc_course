```python
import os
import h5py
import numpy as np
from matplotlib.colors import LinearSegmentedColormap, BoundaryNorm
from matplotlib.pyplot import matshow

def show(forest):
    colors = [(0, 1, 0), (0, 0, 0), (1, 0, 0)]
    cm = LinearSegmentedColormap.from_list('forest', colors, N=3)
    bins = [-1.1, -0.1, 1]
    norm = BoundaryNorm(boundaries=bins, ncolors=4)
    matshow(forest, interpolation='nearest', origin='upper', cmap=cm)

# load steps
forest_steps_path = os.path.join(os.getcwd(), 'forest_steps')
if not os.path.exists(forest_steps_path):
    try:
        os.makedirs(forest_steps_path)
    except IOError:
        raise Exception("Failed to create the required directory {}".format(forest_steps_path))

avail_steps = [step for step in os.listdir(forest_steps_path) if 'forest_' in step]
if not avail_steps:
    raise Exception("No available forest simulation steps were found")

steps = [os.path.join(forest_steps_path, step) for step in avail_steps]
for step in steps:
    with h5py.File(step, 'r') as f:
        a_group_key = f['forest']
        forest_trees = a_group_key.get('forest_tress')
        data = forest_trees[()]
        forest = data.reshape(100,100)
        show(forest)
```
